```python
def min_temp(filename):
    f = open(filename)
    data=csv.reader(f)
    header = next(data)
    min_temp = 99
    min_date = ''

    for row in data:
        if row[-2]=='':
            row[-2] = '99'
        if float(row[-2]) < min_temp:
            min_temp = float(row[-2])
            min_date = row[0]

    return min_date, min_temp
```

```python
def min_temp_df(filename):
    df = pd.read_csv(filename, encoding='cp949')
    df2 = df.sort_values(by = df.columns[3])[0:5]
    return df2.iloc[0,0], df2.iloc[0,3]
```

```python
min_temp_df('seoul.csv')
```

```python
def nth_min_temp_df(filename, n):
    df = pd.read_csv(filename, encoding='cp949')
    df2 = df.sort_values(by = df.columns[3])[0:n]
    
    return df2
```

```python
import numpy as np
f = open('seoul.csv')
data=csv.reader(f)
header = next(data)

x = range(1908, 2021)
n = len(x)
y = np.zeros(n)
num = np.zeros(n)

for row in data:
    if row[2] != '':
        year = int(row[0].split('-')[0])
        temp = float(row[2])
        if year >= 1908 and year <= 2020:
            i = x.index(year)
            y[i]+= temp
            num[i] += 1
    x2 = [x[i] for i in range(len(x)) if num[i] > 300]
    y2 = [y[i] for i in range(len(y)) if num[i] > 300]
y = y/num
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
```

```python
plt.plot(x2, y2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(np.array(x2).reshape(-1,1),y2)
y_pred = model.predict(np.array(x2).reshape(-1,1))
```

```python
plt.plot(x2, y2, 'o')
plt.plot(x2, y_pred)
```
